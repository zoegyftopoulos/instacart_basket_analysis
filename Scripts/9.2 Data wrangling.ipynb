# Create new variables

### Create region variable

```python
# Create lists to organize states by region
northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'New Jersey', 'New York', 'Pennsylvania', 'Rhode Island', 'Vermont' ]
```

```python
midwest = ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']
```

```python
south = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
```

```python
west = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
```

```python
# Use loc() function to create a new region column assigning states to regions
df.loc[df['state'].isin(northeast), 'region'] = 'Northeast'
df.loc[df['state'].isin(midwest), 'region'] = 'Midwest' 
df.loc[df['state'].isin(south), 'region'] = 'South'
df.loc[df['state'].isin(west), 'region'] = 'West'
```

```python
len(df['region'])
```

```python
len(df)
```

```python
df['region'].value_counts(dropna = False)
```

### Create age groups

```python
# Create an age profiling variable
df.loc[(df['age'] >= 18) & (df['age'] < 40), 'age_group'] = 'Young adult'
df.loc[(df['age'] >= 40) & (df['age'] < 60), 'age_group'] = 'Middle-aged adult'
df.loc[df['age'] >= 60, 'age_group'] = 'Old adult'
```

```python
len(df['age_group'])
```

```python
df['age_group'].value_counts(dropna = False)
```

### Create income variable

```python
# Create an income profiling variable
df.loc[df['income'] < 50000, 'income_group'] = 'Low income'
df.loc[(df['income'] >= 50000) & (df['age'] < 100000), 'income_group'] = 'Middle income'
df.loc[df['income'] >= 100000, 'income_group'] = 'High income'
```

```python
len(df['income_group'])
```

```python
df['income_group'].value_counts(dropna = False)
```

### Create a number of dependents variable

```python
# Create a no. of dependents profiling variable
df.loc[df['n_dependants'] == 0, 'dep_group'] = 'no_dependents'
df.loc[df['n_dependants'] > 0, 'dep_group'] = 'has_dependents'
```

```python
len(df['dep_group'])
```

```python
df['dep_group'].value_counts(dropna = False)
```
