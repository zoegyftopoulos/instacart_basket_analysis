#### Aggregating with transform()

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

#### Deriving columns with loc()

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

#### Tasks

#### Check differences in average product prices based on customer loyalty

```python
# Check the basic statistics of the product prices for each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean','min','max']})
```

New customers spend around 3$ more than loyal customers. 

#### Target different types of spenders by creating a spending flag

```python
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.head()
```

#### Determine frequent versus non-frequent customers

```python
ords_prods_merge['median_days_betw_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_betw_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_betw_orders'] > 10) & (ords_prods_merge['median_days_betw_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_betw_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
